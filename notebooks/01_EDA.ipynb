# Exploratory Data Analysis

Load and explore your dataset before training.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Load your dataset
df = pd.read_csv('../data/dataset.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Data types and missing values
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
# Target distribution
target_col = 'target'  # Update this
if target_col in df.columns:
    df[target_col].value_counts().plot(kind='bar')
    plt.title('Target Distribution')
    plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlations')
plt.tight_layout()
plt.show()
```

```python
# Feature distributions
num_cols = df.select_dtypes(include=[np.number]).columns[:6]
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
for i, col in enumerate(num_cols):
    ax = axes[i // 3, i % 3]
    df[col].hist(bins=30, ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```
